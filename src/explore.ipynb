```python
# Your code here
import pandas as pd

df_test = pd.read_csv('https://raw.githubusercontent.com/bartowen/desicion_tree_algo/refs/heads/main/data/processed/clean_test.csv')
df_train = pd.read_csv('https://raw.githubusercontent.com/bartowen/desicion_tree_algo/refs/heads/main/data/processed/clean_train.csv')

df_test.head()
```

```python
df_train.head()
```

```python
X_train = df_train.drop(["Outcome"], axis = 1)
y_train = df_train["Outcome"]
X_test = df_test.drop(["Outcome"], axis = 1)
y_test = df_test["Outcome"]
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state = 42)
model.fit(X_train,y_train)
```

```python
#Graficar los primeros 4 arboles 

import matplotlib.pyplot as plt
from sklearn import tree

fig, axis = plt.subplots(2, 2, figsize = (15, 15))

# Mostramos los 4 primeros árboles de los 100 generados (por defecto)
tree.plot_tree(model.estimators_[0], ax = axis[0, 0], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model.estimators_[1], ax = axis[0, 1], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model.estimators_[2], ax = axis[1, 0], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model.estimators_[3], ax = axis[1, 1], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)

plt.show()
```
